```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-1, 1, 50)
y_1 = 2*x+1
y_2 = x**2

plt.figure(num=1, figsize=(8, 5))
plt.plot(x, y_1)

plt.figure(num=3, figsize=(8, 5))
plt.plot(x, y_2)
plt.plot(x, y_1, color='red', linewidth=1.0, linestyle='--')
plt.show()
```

```python
x = np.linspace(-3, 3, 50)
y_1 = 2*x+1
y_2 = x**2

plt.figure(figsize=(10, 7))
l_1, = plt.plot(x, y_2, label='up', color='black')
l_2, = plt.plot(x, y_1, color='red', linewidth=1.0, linestyle='--', label='down')
plt.legend(handles=[l_1, l_2], labels=['Yao', 'Zhihao'], loc='best')  # 图例

plt.xlim((-1, 2))
plt.ylim((-2, 3))

new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)

plt.yticks([-2, -1.5, -1, 1.22, 3],
           [r'$very\ bad$', r'$bad$', r'$normal$', r'$good$', r'$very\ good$'])

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', -1))

plt.show()
```

```python
X = np.linspace(-3, 3, 50)
Y = 2*X+1

plt.figure(figsize=(10, 7))
plt.plot(X, Y)

plt.xlim((-3, 3))
plt.ylim((-6, 8))
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

x_0 = -1
y_0 = 2*x_0+1
plt.scatter(x_0, y_0, s=50, color='b')
plt.plot([x_0, x_0], [y_0, 0], color='red', linestyle='--', linewidth=2.5)

# method 1
##################################
plt.annotate(r'$\mu_1(%s,%d)$' % (x_0, y_0), xy=(x_0, y_0), xycoords='data',
             xytext=(+10, -80), textcoords='offset points', fontsize=10,
             arrowprops=dict(arrowstyle='<-', connectionstyle='arc3,rad=0.2'))
# xy 数据
# xycoords基于‘data’ 箭头的指向
# xytext文本偏移量，textcoords文本偏移方式，fontsize文本字体
# arrowprops （arrowstyle箭头, connectionstyle箭头的弧度）
##################################
# method 2
##################################
plt.text(-2.5, 2, r'$This\ is\ the\ some\ text.\ \mu\ \alpha_i\ \sigma_j$',
         fontdict={'size': 13, 'color': 'blue'})
##################################
plt.show()
```

```python
n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
T = np.arctan2(Y, X)  # for color value

plt.scatter(X, Y, c=T, alpha=0.5)  # 散点图scatter

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.xticks(())
plt.yticks(())

plt.show()
```

```python
n = 12
X = np.arange(n)
Y_1 = (1-X/float(n))*np.random.uniform(0.5, 1.0, n)
Y_2 = (1-X/float(n))*np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y_1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y_2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, Y_1):
    # ha:horizontal alignment 对齐方式
    plt.text(x, y+0.05, '%.2f' % y, ha='center', va='bottom')
for x, y in zip(X, Y_2):
    # ha:horizontal alignment
    plt.text(x, -y-0.05, '%.2f' % y, ha='center', va='top')

plt.xlim(-0.5, n)
plt.ylim(-1.25, 1.25)
plt.xticks(())
plt.yticks(())

plt.show()
```

```python
def f(a, b):
    # the height function
    return (1-a/2+a**5+b**3)*np.exp(-a**2-b**2)


n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

# use plt.contourf to filling contour
# 1. X,Y and value for(X, Y) point
plt.contourf(X, Y, f(X, Y), 8, alpha=0.75, cmap=plt.cm.hot)

# 2. use plt.contour to add contour lines
C = plt.contour(X, Y, f(X, Y), colors='black', linewidth=0.5)

# 3. adding label
plt.clabel(C, inline=True, fontsize=10)

plt.xticks(())
plt.yticks(())
plt.show()
```

```python
fig,ax=plt.subplots()

x = np.arange(0,2*np.pi,0.01)
line,=ax.plot(x,np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x+i/10))
    return line,

def init():
    line.set_ydata(np.sin(x))
    return line,

ani = animation.FuncAnimation(fig=fig,func=animate,frames=100,init_func=init,interval=20,blit=True)
plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
# X,Y value
X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2 + Y**2)
# height value
Z = np.sin(R)

ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow'))
ax.contourf(X,Y,Z,zdir='z',offset = -2,cmap='rainbow')
ax.set_zlim(-2,2)

plt.show()
```
